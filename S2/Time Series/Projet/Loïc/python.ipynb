```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"]=30,10
```

```python
series = pd.read_csv('data.csv', header=0, index_col=0)
series.plot()
plt.show()
```

```python
X = series.values
diff = [X[i] - X[i - 365] for i in range(365, len(X))]
```

```python
plt.plot(diff)
plt.show()
```

```python
y = series.values[365:] - diff
```

```python
plt.plot(y)
plt.show()
```
